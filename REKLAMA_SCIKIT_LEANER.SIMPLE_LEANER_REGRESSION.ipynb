```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/reklama_sales.csv')
df.head()
```

```python
sns.pairplot(df[['TV','Radio','Newspaper','Sales']])
plt.show()
```

```python
x = df['TV'].to_numpy()
x
```

```python
y = df['Sales'].to_numpy()
y
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=df, x='TV', y='Sales')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.regplot(data=df, x='TV', y='Sales', line_kws={'color':'red'})
plt.show()
```

```python
# 1-chi
```

```python
xmean = np.mean(x)

ymean = np.mean(y)
```

```python
theta1 = sum((x-xmean)*(y-ymean))/sum((x-xmean)**2)
print(f"{theta1=}")
```

```python
theta0 = ymean-theta1*xmean
print(f"{theta0=}")
```

```python
# 2-chi
```

```python
x_test = df.sample(10, random_state=42)['TV'].to_numpy()
print(f"{x_test=}")
```

```python
x_test = df.sample(10, random_state=42)['TV'].to_numpy()
print(f"{x_test=}")
y_test = df.sample(10, random_state=42)['Sales'].to_numpy()
print(f"{y_test=}")
```

```python
y_predict = theta0 + theta1*x_test
print(f"{y_predict=}")
```

```python
# 3-chi
```

```python
# MAE

mae = np.sum(np.absolute(y_predict-y_test))/len(y_test)
print(f"{mae=}")
```

```python
# RMSE

rmse = np.sqrt(np.sum((y_predict-y_test)**2)/len(y_test))
print(f"{rmse=}")
```

```python
# linear regression
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.1, random_state=42)
```

```python
from sklearn import linear_model
lr_model = linear_model.LinearRegression()

x_train = np.asanyarray(train_set[['TV']])
y_train = np.asanyarray(train_set[['Sales']])

#modelni tartiblaymiz ( train )

lr_model.fit(x_train, y_train)
```

```python
# theta1 va theta0 koefisentlarni ajratib olamiz

theta1 = lr_model.coef_[0][0]
theta0 = lr_model.intercept_[0]
print(f"{theta1=}")
print(f"{theta0=}")
```

```python
# theta0 va theta1 qiymatlar yordamida togri chiziq chizamiz
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='TV', y='Sales')
plt.plot(x_train, theta1*x_train + theta0,'-r')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```

```python
# modelni baholaymiz
```

```python
x_test = np.asanyarray(test_set[['TV']])

y_test = np.asanyarray(test_set[['Sales']])
```

```python
y_predict = lr_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_test, y_predict)

mrse = mean_squared_error(y_test, y_predict)

print(f"{mae=}")
print(f"{mrse=}")
```
